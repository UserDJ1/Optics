```julia
using Plots
gr()
using LaTeXStrings
```

```julia
function Rperp()
    return (sin( - asin((1/1.5)*sin()))/sin( + asin((1/1.5)*sin())))^2
end

function Tperp()
    return (1.5*cos(asin((1/1.5)*sin()))/cos())*(((2*sin(asin((1/1.5)*sin()))*cos())/(sin( + asin((1/1.5)*sin())))))^2
end

t = collect(range(0, stop=pi/2, step=0.001))

plot(t, Rperp.(t), label=L"R_{\!\perp}")
plot!(t, Tperp.(t), label=L"T_{\!\perp}")
plot!(xlim=(0,pi/2), ylim=(0,1), ylabel="Reflectance and \ntransmittance", xlabel=L"\theta_i \;\;\mathrm{ (degrees)}")
```

```julia
function Rpar()
    return ((1.5*cos() - cos(asin((1/1.5)*sin())))/(cos(asin((1/1.5)*sin())) + 1.5*cos()))^2
end

function Tpar()
    return (1.5*cos(asin((1/1.5)*sin()))/cos())*((2*cos())/(cos(asin((1/1.5)*sin())) + 1.5*cos()))^2
end

t = collect(range(0, stop=pi/2, step=0.001))

plot(t, Rpar.(t), label=L"R_{\!\parallel}")
plot!(t, Tpar.(t), label=L"T_{\!\parallel}")
plot!(xlim=(0,pi/2), ylim=(0,1), ylabel="Reflectance and \ntransmittance", xlabel=L"\theta_i \;\;\mathrm{ (degrees)}")
```
