```julia
using Plots
gr()
using LaTeXStrings
```

```julia
function Rperp(θ)
    return (sin(θ - asin((1/1.5)*sin(θ)))/sin(θ + asin((1/1.5)*sin(θ))))^2
end

function Tperp(θ)
    return (1.5*cos(asin((1/1.5)*sin(θ)))/cos(θ))*(((2*sin(asin((1/1.5)*sin(θ)))*cos(θ))/(sin(θ + asin((1/1.5)*sin(θ))))))^2
end

t = collect(range(0, stop=pi/2, step=0.001))

plot(t, Rperp.(t), label=L"R_{\!\perp}")
plot!(t, Tperp.(t), label=L"T_{\!\perp}")
plot!(xlim=(0,pi/2), ylim=(0,1), ylabel="Reflectance and \ntransmittance", xlabel=L"\theta_i \;\;\mathrm{ (degrees)}")
```

```julia
function Rpar(θ)
    return ((1.5*cos(θ) - cos(asin((1/1.5)*sin(θ))))/(cos(asin((1/1.5)*sin(θ))) + 1.5*cos(θ)))^2
end

function Tpar(θ)
    return (1.5*cos(asin((1/1.5)*sin(θ)))/cos(θ))*((2*cos(θ))/(cos(asin((1/1.5)*sin(θ))) + 1.5*cos(θ)))^2
end

t = collect(range(0, stop=pi/2, step=0.001))

plot(t, Rpar.(t), label=L"R_{\!\parallel}")
plot!(t, Tpar.(t), label=L"T_{\!\parallel}")
plot!(xlim=(0,pi/2), ylim=(0,1), ylabel="Reflectance and \ntransmittance", xlabel=L"\theta_i \;\;\mathrm{ (degrees)}")
```
